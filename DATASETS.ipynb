```python
def convert_num(val):
    """
    Convert the string number value to a float
     - Remove commas
     - Convert to float type
    """
    try:
        new_val = val.replace(',','')
    except:
        new_val = val
    return float(new_val)
```

```python
def convert_percent(val):
    """
    Convert the string percent value to a float
     - Remove %
     - Divedes 100
     - Convert to float type
    """
    try:
        new_val = val.replace('%','')
    except:
        new_val = val
    return float(new_val)/100
```

```python
def convert_time(val):
    """
    Convert the string time value to a float for seconds
     - Remove :
     - Convert to minutes
     - Convert to float type
    """
    try:
        time = val.split(":")
        new_val = float(time[0])*60 + float(time[1])
        
    except:
        new_val = val
    return new_val
```
